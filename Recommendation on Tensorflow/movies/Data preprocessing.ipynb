```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

# 处理得到用户评分矩阵ratings

```python
ratings_df = pd.read_csv("ml-latest-small/ratings.csv")
```

```python
movie_df = pd.read_csv("ml-latest-small/movies.csv")
```

```python
movie_df.tail()
```

```python
movie_df["movieRow"] = movie_df.index
```

```python
ratings_df = pd.merge(ratings_df,movie_df,on='movieId')
```

```python
ratings_df['movieId'] = ratings_df['movieRow']
```

```python
ratings = ratings_df[['userId','movieId','rating']]
```

```python
ratings.tail()
```

```python
ratings.to_csv('ratingsProcessed.csv',index=False,header=True,encoding='utf-8')
```

# 将用户评分矩阵ratings分为训练集train_rating和测试集test_rating

```python
x,y = ratings[['userId','movieId']],ratings[['rating']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
train_rating = pd.merge(x_train,y_train,right_index=True,left_index=True)
```

```python
test_rating = pd.merge(x_test,y_test,right_index=True,left_index=True)
```

```python
train_rating.to_csv('train_rating.csv',index=False,header=True,encoding='utf-8')
```

```python
test_rating.to_csv('test_rating.csv',index=False,header=True,encoding='utf-8')
```
