```python
import pandas as pd
import numpy as np
import pymysql
```

```python
conn=pymysql.connect(host="localhost",user="root",passwd="liugang666",db="goodbooks",charset="utf8")
```

```python
query_bookRatings_sql = "select * from ratings where user_id=5 and rating=5"
```

```python
ratings_df = pd.read_sql(query_bookRatings_sql,con=conn)
```

```python
ratings_df.head()
```

```python
booksList = ratings_df["book_id"].tolist()
```

```python
booksList
```

```python
read_books = pd.DataFrame()
for bookid in booksList:
    qsql = "select original_title as title  ,`authors` ,image_url from books where book_id="+bookid
    book_info = pd.read_sql(qsql,con=conn)
    read_books = read_books.append(book_info,ignore_index=True)
```

```python
read_books
```

```python
read_books.to_csv("read_books.csv",index=False,header=True,encoding='utf-8')
```
