```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
ratings_df = pd.read_csv("ratings.csv")
```

```python
books_df = pd.read_csv("books.csv")
```

```python
rating_max = ratings_df["rating"].max()
```

```python
rating_min = ratings_df["rating"].min()
```

```python
rating_max,rating_min
```

```python
booksId_max = books_df["book_id"].max()
```

```python
booksId_max
```

```python
userId_max = ratings_df["user_id"].max()
```

```python
userId_max
```

```python
selectUserId = int(input("请输入你要推荐的用户ID："))
```

```python
selectUserId 
```

```python
selectUser_bookList = ratings_df.loc[ratings_df["user_id"]==selectUserId]["book_id"].values
```

```python
selectUser_bookList
```

```python
selectUser_bookList.shape
```

```python
simUserCount = np.zeros(userId_max,dtype=np.int)
```

```python
simUserCount.shape
```

```python
print(simUserCount)
```

```python
for index,row in ratings_df.iterrows():
    if row["book_id"] in selectUser_bookList:
        simUserCount[row["user_id"]] = simUserCount[row["user_id"]]+1
```

```python
simUserCount
```

```python
simUserId = np.argsort(-simUserCount)
```

```python
trainUserId = simUserId[0:50]
```

```python
trainUserId = trainUserId.tolist()
```

```python
trainUserId
```

```python
trainBookId = []
```

```python
for index,row in ratings_df.iterrows():
    if row["user_id"] in trainUserId and row["book_id"]  not in trainBookId:
        trainBookId.append(row["book_id"])
```

```python
trainBookId
```

```python
trainUserNo,trainBookNo = len(trainUserId),len(trainBookId)
```

```python
trainUserNo,trainBookNo
```

```python
train_ratings = np.zeros((trainUserNo,trainBookNo),dtype=np.float32)
```

```python
for index,row in ratings_df.iterrows():
    if row["user_id"] in trainUserId and row["book_id"] in trainBookId:
        train_ratings[trainUserId.index(row["user_id"])][trainBookId.index(row["book_id"])] = row["rating"]
```

```python
train_ratings
```

```python
train_record = train_ratings>0
```

```python
train_record = np.array(train_record,dtype=int)
```

```python
train_record
```

```python
temSum = 0.0
train_count = 0
for i in range(0,trainUserNo):
    for j in range(0,trainBookNo):
        if train_ratings[i][j] != 0. :
            temSum = temSum + train_ratings[i][j]
            train_count = train_count+1
train_global_rating_mean = temSum/train_count
```

```python
train_global_rating_mean
```

```python
f = 10
lamda = 0.01
```

```python
def createBSVD(mean,userNo,bookNo,f):
    BIAS_u = train_global_rating_mean
    BIAS_bu = tf.Variable(tf.zeros((userNo,1),dtype=tf.float32))
    BIAS_bi = tf.Variable(tf.zeros((1,bookNo),dtype=tf.float32))
    BIAS_P = tf.Variable(tf.random_normal([userNo,f],stddev=1/(f**0.5),dtype=tf.float32))
    BIAS_Q = tf.Variable(tf.random_normal([bookNo,f],stddev=1/(f**0.5),dtype=tf.float32))
    BIAS_predict_rating = BIAS_u + BIAS_bu + BIAS_bi + tf.matmul(BIAS_P,BIAS_Q,transpose_b=True)
    BIAS_loss = tf.reduce_sum(((train_ratings-BIAS_predict_rating)*train_record)**2) + lamda*((tf.reduce_sum(BIAS_P)**2) + (tf.reduce_sum(BIAS_Q)**2) + (tf.reduce_sum(BIAS_bu)**2) + (tf.reduce_sum(BIAS_bi)**2))
    return BIAS_loss,BIAS_predict_rating
```

```python
loss,predict = createBSVD(train_global_rating_mean,trainUserNo,trainBookNo,f)
```

```python
learning_rate = 1e-4
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
```

```python
train = optimizer.minimize(loss)
```

```python
sess = tf.Session()
```

```python
init = tf.global_variables_initializer()
```

```python
sess.run(init)
```

```python
for i in range(1000): 
    if i % 100 == 0:
        print(i)
    sess.run(train)
```

```python
predict_rating = sess.run(predict)
```

```python
predict_rating
```

```python
selectUser_predictRating = predict_rating[0]
```

```python
selectUser_predictRating
```

```python
selectUser_sortRating = np.argsort(-selectUser_predictRating)
```

```python
selectUser_sortRating
```

```python
predict_rating_sort = np.sort(selectUser_predictRating)[::-1]
```

```python
predict_rating_sort
```

```python
predict_bookId = []
for i in range(0,trainBookNo):
    predict_bookId.append(trainBookId[selectUser_sortRating[i]])
```

```python
predict_bookId
```

```python
recomResult = {}
for i in range(0,20):
    recomResult[predict_bookId[i]] = predict_rating_sort[i]
```

```python
recomResult.keys()
```

```python
recomlist = recomResult.keys()
```

```python
for bid in recomlist:
    if bid in selectUser_bookList:
        recomResult.pop(bid)
```

```python
books_df.head()
```

```python
print("为用户ID为：%d的用户推荐的10本书是：" % (selectUserId))
idx = 1
for bid,prating in recomResult.items():
    bookname = books_df.loc[books_df["book_id"]==bid]["original_title"].values[0]
    print("第 %d 本图书id为  %d ，书名为：%s ，评分为：%.2f" % (idx,bid,bookname,prating))
    if idx == 10:
        break
    idx = idx +1
```
